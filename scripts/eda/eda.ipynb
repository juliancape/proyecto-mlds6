# Análisis exploratorio de los datos

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('/work/diabetes_prediction_dataset.csv')
```

### 1. Exploración dataframe

```
df.head()
```

### 2. Datos categóricos 

```
print(df['gender'].unique())
print(df['smoking_history'].unique())
```

### 3. Descripción datos numericos

```
df.describe()
```

### 4. Datos faltantes

```
print(df.isnull().sum())
```

### 5. Análisis variables

Age - edad

```
plt.hist(df['age'], bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

bmi- porcentaje grasa corporal

```
sns.displot(df['bmi'], bins=30)
plt.title('BMI Distribution')
plt.show()
```

Variables binarias

```
for col in ['hypertension', 'heart_disease', 'diabetes']:
    sns.countplot(x=col, data=df)
    plt.title(f'{col} Distribution')
    plt.show()
```

smoking_history - persona fumadora

```
sns.countplot(x='smoking_history', data=df)
plt.title('Smoking History Distribution')
plt.show()
```

correlacion del dataframe

```
heatmap = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```
